Library Import

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm
import ipywidgets as widgets
from IPython.display import display
from IPython import get_ipython
```

Test Z-score
Z-score là cách đo một điểm dữ liệu nằm cách giá trị trung bình bao nhiêu độ lệch chuẩn (standard deviations).


```python
# %pip install ipywidgets 
x_slider = widgets.FloatSlider(
    value=8.5,
    min=-15,
    max=15,
    step=0.1,
    description='x'
)

mu_slider = widgets.FloatSlider(
    value=4,
    min=-10,
    max=10,
    step=0.1,
    description='μ'
)

sigma_slider = widgets.FloatSlider(
    value=1,
    min=0.1,
    max=5,
    step=0.1,
    description='σ'
)

formula = widgets.HTML(
"""
    <div style="
        font-size:48px;
        text-align:center;
        color:white;
        margin-bottom:10px;
    ">
        z = (x - μ) / σ
    </div>
"""
)

result_html = widgets.HTML()

plot_output = widgets.Output()

def update(*args):
    x = x_slider.value
    mu = mu_slider.value
    sigma = sigma_slider.value

    z = (x - mu) / sigma
    prob = norm.cdf(z)

    result_html.value = f"""
    <div style='font-size:18px'>
        <b>z</b> = ({x:.2f} - {mu:.2f}) / {sigma:.2f}
        = <b>{z:.3f}</b>
        <br><br>
        Φ(z) ≈ <b>{prob:.2%}</b>
        <br><br>
            x: giá trị cần kiểm tra<br>
            μ: giá trị trung bình (mean)<br>
            σ: độ lệch chuẩn (standard deviation)<br>
            z: số độ lệch chuẩn mà điểm đó cách mean
    </div>
    """

    with plot_output:
        plot_output.clear_output(wait=True)

        xs = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
        ys = norm.pdf(xs, mu, sigma)

        fig, ax = plt.subplots(figsize=(6,4))

        ax.plot(xs, ys, linewidth=2)
        ax.fill_between(xs, ys, alpha=0.25)

        ax.axvline(x, linewidth=2)

        ax.set_title("Normal Distribution")
        ax.grid(alpha=0.3)

        plt.show()

for w in [x_slider, mu_slider, sigma_slider]:
    w.observe(update, names='value')

left_panel = widgets.VBox([
    formula,
    x_slider,
    mu_slider,
    sigma_slider,
    result_html
])

ui = widgets.HBox([
    left_panel,
    plot_output
])

get_ipython().run_line_magic('matplotlib', 'widget')
update()
display(ui)
```
